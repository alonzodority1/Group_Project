```python
# Generate our categorical variable lists
categories = ["Soybeans_Open", "Soybeans_High", "Soybeans_Low", "Soybeans_Close", "Soybeans_Volume",
              "Corn_Open", "Corn_High", "Corn_Low", "Corn_Close", "Corn_Volume",
              "Wheat_Open", "Wheat_High", "Wheat_Low", "Wheat_Close", "Wheat_Volume",
              "Cotton_Open", "Cotton_High", "Cotton_Low", "Cotton_Close", "Cotton_Volume",
              "Crude_Open", "Crude_High", "Crude_Low", "Crude_Close", "Crude_Volume",
              "USDBRL_Open", "USDBRL_High", "USDBRL_Low", "USDBRL_Close", "USDBRL_Volume",
              "EURUSD_Open", "EURUSD_High", "EURUSD_Low", "EURUSD_Close", "EURUSD_Volume",
              "IR_Open", "IR_High", "IR_Low", "IR_Close", "IR_Volume"] 
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(alldata_df[categories])) 

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(categories)
encode_df.head() 
```

```python
# Merge one-hot encoded features and drop the originals
alldata_df = alldata_df.merge(encode_df, left_index= True, 
                                      right_index= True).drop(columns= categories,axis= 1)
alldata_df.head()  
```

```python
# Split our preprocessed data into our features and target arrays
X = alldata_df.drop("Soybeans_Close", axis= 1)
y = alldata_df.TARGET_COLUMN_NAME

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 0, stratify= y)  
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test) 
```

# Train the Neural Network 

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_layer_1 = 100
hidden_layer_2 = 50

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer_1, activation= "relu",
                             input_dim= number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units= hidden_layer_2, activation= "relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units= 1, activation= "sigmoid")) 

# Check the structure of the model
nn.summary() 
```

```python
# Compile the model
nn.compile(loss= "LOSS_FUNCTION", optimizer= "adam", metrics= ["accuracy"]) 
```

```python
# Train the model
model = nn.fit(X_train_scaled, y_train, epochs= 100) 
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}") 
```
